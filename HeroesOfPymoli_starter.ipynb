```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = r"C:\Users\mengc\Desktop\Pandas-Challenge\Instructions\HeroesOfPymoli\Resources\purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
playerData = purchase_data.loc[:, ["Gender", "SN", "Age"]]
playerData = playerData.drop_duplicates()
totalPlayers = playerData.count()[0]
#totalPlayers
pd.DataFrame({"Total Players": [totalPlayers]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
meanPrice = purchase_data["Price"].mean()
sumPrice = purchase_data["Price"].sum()
countPrice = purchase_data["Price"].count()
countItem = len(purchase_data["Item ID"].unique())

table = pd.DataFrame({"Number of unique items": countItem,
                      "Total revenue": [sumPrice],
                      "Number of purchases": [countPrice],
                      "Average price": [meanPrice]
                     })

table = table.round(2)
table["Total revenue"] = table["Total revenue"].map("${:,.2f}".format)
table["Number of purchases"] = table["Number of purchases"].map("{:,}".format)
table["Average price"] = table["Average price"].map("${:,.2f}".format)

table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
totalGender = playerData["Gender"].value_counts()
pctGender = totalGender/totalPlayers
genderTable = pd.DataFrame({"Total count": totalGender,
                            "Percentage of Players": pctGender})

genderTable["Percentage of Players"] = genderTable["Percentage of Players"].map("{:,.2%}".format)
genderTable[{"Total count", "Percentage of Players"}]

genderTable
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genderTotalPurchase = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Gender Total purchase value")
genderAvgPurchase = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Gender Average purchase value")
genderCount = purchase_data.groupby(["Gender"]).count()["Price"].rename("Gender Count")

genderTotalPerPerson = genderTotalPurchase / genderTable["Total count"]

genderDataTable = pd.DataFrame({"Purchase Count": genderCount,
                    "Average Purchase Price": genderAvgPurchase,
                    "Total Purchase Value": genderTotalPurchase,
                    "Average Purchase Total per Person by Gender": genderTotalPerPerson})

genderDataTable["Purchase Count"] = genderDataTable["Purchase Count"].map("{:,}".format)
genderDataTable["Average Purchase Price"] = genderDataTable["Average Purchase Price"].map("${:,.2f}".format)
genderDataTable["Total Purchase Value"] = genderDataTable["Total Purchase Value"].map("${:,.2f}".format)
genderDataTable["Average Purchase Total per Person by Gender"] = genderDataTable["Average Purchase Total per Person by Gender"].map("${:,.2f}".format)
genderDataTable[{"Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Purchase Total per Person by Gender"}]

genderDataTable
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ageBins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 9999]
ageGroups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

playerData["Age Ranges"] = pd.cut(playerData["Age"], ageBins, labels=ageGroups)
ageTotal = playerData["Age Ranges"].value_counts()
agePercent = ageTotal/totalPlayers

ageTable = pd.DataFrame({"Total Count": ageTotal,
                         "Percentage of Players": agePercent})

ageTable["Total Count"] = ageTable["Total Count"].map("{:,}".format)
ageTable["Percentage of Players"] = ageTable["Percentage of Players"].map("{:,.2%}".format)

ageTable = ageTable.sort_index()
ageTable
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], ageBins, labels=ageGroups)

agePurchaseSum = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
agePurchaseMean = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
agePurchaseCount = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

ageTotalPerPerson = agePurchaseSum / ageTotal

agePurchaseData = pd.DataFrame({"Purchase Count": agePurchaseCount,
                       "Average Purchase Price": agePurchaseMean,
                       "Total Purchase Value": agePurchaseSum,
                       "Average Purchase Total per Person by Age Group": ageTotalPerPerson})
agePurchaseData
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
userSum = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase value")
userMean = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase value")
userCount = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")

userData = pd.DataFrame({"Total Purchase Value": userSum,
                         "Average Purchase Value": userMean,
                         "Purchase Count": userCount})


userSort = userData.sort_values("Total Purchase Value", ascending=False)


userSort[["Purchase Count", "Average Purchase Value", "Total Purchase Value"]].head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
itemData = purchase_data[["Item ID", "Item Name", "Price"]]

itemSum = itemData.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
itemMean = itemData.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")
itemCount = itemData.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

itemData = pd.DataFrame({"Total Purchase Value": itemSum,
                         "Item Price": itemMean,
                         "Purchase Count": itemCount})
itemData
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
itemDataSorted = itemData.sort_values("Purchase Count", ascending = False)
itemDataSorted[["Purchase Count", "Item Price", "Total Purchase Value"]].head(5)
```
